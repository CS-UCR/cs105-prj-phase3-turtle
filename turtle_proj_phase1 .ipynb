```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd

real_name = [] 
name = []
size = []
animal = [] 
```

```python
page = requests.get("https://www.montereybayaquarium.org/animals-and-exhibits/animal-guide/birds")
soup = BeautifulSoup(page.content, 'html.parser')

links_with_text = []

for a in soup.find_all('a',href=True,class_='image-grid-item'):
    if a.text:
        links_with_text.append(a['href'])      
        
links_count += len(links_with_text)

for x in links_with_text:
    page = requests.get(x)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    facts_tags = soup.select('.ag-facts--left p')
    facts = [x.get_text() for x in facts_tags]
    
    real_name.append(facts[0]) 
    animal.append(facts[1])
    size.append(facts[3])
    
    name_tags = soup.select('.page-title')
    name_text = [x.getText() for x in name_tags]
    name.append(name_text)
```
